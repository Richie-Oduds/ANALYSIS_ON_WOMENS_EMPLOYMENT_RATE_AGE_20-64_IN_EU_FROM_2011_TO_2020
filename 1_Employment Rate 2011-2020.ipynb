#### QUESTION 1:
##### What is the average employment rate of women in the EU from 2011-2020?

```python
employment_rate.head()
```

```python
employment_rate_average = employment_rate[["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]].mean()
```

```python
employment_rate_average.round(2)
```

```python
#Converting the employment_rate_average for each year into pandas dataframe for easy save into excel file
year_2011 = [61.06]
year_2012 = [61.36]
year_2013 = [61.61]
year_2014 = [62.57]
year_2015 = [63.61]
year_2016 = [64.67]
year_2017 = [66.11]
year_2018 = [67.29]
year_2019 = [68.19]
year_2020 = [67.31]
df = pd.DataFrame([year_2011, year_2012, year_2013, year_2014, year_2015, year_2016, year_2017, year_2018, year_2019, year_2020]) 
df = df.transpose() 
df.columns = ["year_2011", "year_2012", "year_2013", "year_2014", "year_2015", "year_2016", "year_2017", "year_2018", "year_2019", "year2020"] 
df.head()
```

#### QUESTION 2:
##### How is the rate of change for women employment in the EU from 2011-2020

To find the rate of change for each year from 2011 to 2020, we have to find the total percentage sum for each year.

```python
#To find the total percent sum for each year, we have to drop the string column in order for our syntax to read our data
employment_rate_drop = employment_rate.drop(columns=['Geographic region\Time','2009','2010'])
employment_rate_drop.head()
```

```python
#Total percent sum for each year
total_percent_sum = employment_rate_drop.sum(axis=0)
total_percent_sum
```

```python
#Converting total percent sum for each year into pandas dataframe for easy save into excel file
year_2011 = [1892.8]
year_2012 = [1902.3]
year_2013 = [1910.0]
year_2014 = [1939.7]
year_2015 = [1972.0]
year_2016 = [2004.8]
year_2017 = [2049.4]
year_2018 = [2086.0]
year_2019 = [2113.9]
year_2020 = [2086.7]
data = pd.DataFrame([year_2011, year_2012, year_2013, year_2014, year_2015, year_2016, year_2017, year_2018, year_2019, year_2020]) 
data = data.transpose() 
data.columns = ["year_2011", "year_2012", "year_2013", "year_2014", "year_2015", "year_2016", "year_2017", "year_2018", "year_2019", "year2020"] 
data.head()
```

```python
percent_sum = data
percent_sum.columns = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
percent_sum
```

```python
# Save dataframe to Excel Workbook
percent_sum.to_excel("total_percent_sum_2011_to_2020.xlsx", index = False)
```

Next, is to find the rate of change for women employment in the EU from 2011-2020

```python
#The change in employment rate from 2011 to 2020
total_percent_sum.pct_change().round(3)
```

```python
#Converting the change in years into pandas dataframe for easy save into excel file
year_2012 = [0.005]
year_2013 = [0.004]
year_2014 = [0.016]
year_2015 = [0.017]
year_2016 = [0.017]
year_2017 = [0.022]
year_2018 = [0.018]
year_2019 = [0.013]
year_2020 = [-0.013]
data1 = pd.DataFrame([year_2012, year_2013, year_2014, year_2015, year_2016, year_2017, year_2018, year_2019, year_2020]) 
data1 = data1.transpose() 
data1.columns = ["year_2012", "year_2013", "year_2014", "year_2015", "year_2016", "year_2017", "year_2018", "year_2019", "year2020"] 
data1.head()
```

```python
change = data1
change.columns = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
```
